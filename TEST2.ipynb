```
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
```

```
X = np.arange(-5.0, 5.0, 0.1)


Y = 1-4/(1+np.power(3, X-2))

plt.figure(figsize=(8,5))
plt.plot(X,Y) 
plt.ylabel('Dependent Variable')
plt.xlabel('Indepdendent Variable')
plt.show()
```
